## Drawing on Images 

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2

# If running in Jupyter Notebook, uncomment the line below
#%matplotlib inline  
```

```python
import matplotlib
print(matplotlib.__file__)
```

```python
# Making the 3D array of ones function 
np.ones(shape=(3,3,3) , dtype =float)
```

```python
# multipliting by 255 (for white background) and for the black background use 0 
blank_image = np.ones(shape=(512,512,3), dtype =np.int8) *255
```

```python
blank_image.shape
```

```python
plt.imshow(blank_image)
plt.show()
```

```python
cv2.rectangle(blank_image,
             pt1 =(200,0),
             pt2 =(85,128),
             color =(200,0,255),
             thickness =2)

plt.imshow(blank_image)
```

###  Circle

```python
cv2.circle(img= blank_image,
          center =(300,300),
          radius =100,
           color =(255,0,0),
           thickness =2)

plt.imshow(blank_image)
```

### Fill the Circle 

```python
cv2.circle(img= blank_image,
          center =(300,300),
          radius =100,
           color =(200,0,255),
           thickness =2)

plt.imshow(blank_image)
```

### Creating The LInes

```python
# Draw a diagonal blue line with thickness of 5 px
cv2.line(blank_image,
         pt1 =(123,124),
         pt2 =(400,400),
         color =(255,255,20),
         thickness =5)

plt.imshow(blank_image)
```

### Creating the Text

```python
# Draw a diagonal blue line with thickness of 5 px

cv2.putText(blank_image,
            text ="Anju",
            org =(10,400),
            fontFace =None,
            fontScale=4,
             color =(12,255,123),
             thickness =5)

plt.imshow(blank_image)
```

### Creating the Canvas 

```python
canvas = np.zeros((1000,1000,3), dtype="uint8")
for i in range(60):

    r= np.random.randint(5,20)
    c= np.random.randint(0,255, size =(3,)).tolist()
    pt =np.random.randint(0,1000,size =(2,))
    cv2.circle(canvas, tuple(pt),r,c,-1)

    pt1 =np.random.randint(0,1000,size =(2,))
    pt2 =np.random.randint(0,1000,size =(2,))
    c= np.random.randint(0,255, size =(3,)).tolist()
    cv2.rectangle(blank_image,
             pt1 =pt1,
             pt2 =pt2,
             color =c,
             thickness =2)


```

```python
cv2 .imshow("tes", canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
