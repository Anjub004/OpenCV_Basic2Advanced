### Rotate Video

```python
def rotate(image , angle, center =None , scale=1.0):

    (h,w) = image.shape[:2]

    if center == None:
        center = (w//2, h//2)

    # Create matrix for rotation
    M = cv2.getRotationMatrix2D(center , angle, scale)
    rotated = cv2.warpAffine(image , M, (w,h))

    return rotated
```

```python
# cap = cv2.VideoCapture(0)
cap = cv2.VideoCapture("DATA/vid2.mp4")

time.sleep(2)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
print(width, height)

angle = 0

while True:

    if angle==360:
        angle = 0

    # Capture frame-by-frame
    ret ,frame = cap.read()
    #---------------------------
    print("ret" , ret)

    if ret == False:
        break

    frame =cv2.resize(frame, None, fx=0.7 ,fy =0.7)

    rframe =rotate(frame, angle)

    angle +=1
    time.sleep(0.06)

    # # Display the resulting frame
    # cv2.imshow("Frame" , frame)
    # #------------------------------
    cv2.imshow("Rotated Frame" , rframe)
    # This command let's us quit with the "q" button on a Keyboard
    if cv2.waitKey(1) & 0xFF ==ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

### Convert the Video to Gray

```python
#cap = cv2.VideoCapture(0)
cap =cv2.VideoCapture("DATA/vid2.mp4")

time.sleep(2)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

while True:

    # Capture the frame- by- frame
    ret , frame =cap.read()

    if ret ==False:
        break

    frame =cv2.resize(frame, None, fx=0.7 ,fy =0.7)


    # Our operations on the frame come here
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    time.sleep(0.006)
    # Display the resulting frame
    cv2.imshow("Gray Frame" , gray)
    cv2.imshow("Original Color", frame)

    if cv2.waitKey(1) & 0xFF ==ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

### Making the Color Channel Frames

```python
#cap = cv2.VideoCapture(0)
cap =cv2.VideoCapture("DATA/vid2.mp4")

time.sleep(2)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

while True:

    # Capture the frame- by- frame
    ret , frame =cap.read()

    if ret ==False:
        break

    image =cv2.resize(frame , None, fx = 0.5, fy= 0.5)
    (B,G,R)  = cv2.split(image)

    z = np.zeros(image.shape[:2], dtype ="uint8")

    # Our operations on the frame come here
    cv2.imshow("RED" , cv2.merge([z,z,R]))
    cv2.imshow("GREEN" , cv2.merge([z,G,z]))
    cv2.imshow("BLUE" , cv2.merge([B,z,z]))

    # time.sleep(0.06)
    # Display the resulting frame
    cv2.imshow("Original Color", frame)

    if cv2.waitKey(1) & 0xFF ==ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
