## Cropping

```python
# Import Dependencies
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the Image 
image = cv2.imread("DATA/DOGS.jpeg")
height ,width = image.shape[:2]
```

```python
# Check type and shape
print(type(image))
print()
print(image.shape)
```

```python
# Lets get the ending pixel coordinates (bottom right)
end_row ,end_col = int(height * .75) , int(width * .75)
```

```python
start_row ,start_col = int(height * .25) , int(width * .25)
```

```python
# Check if the image was loaded successfully
if image is None:
    print("Error: Image not found.")
else:
    # Define the region of interest for cropping
    # Here, (x, y) is the top-left coordinate and (w, h) is the width and height of the rectangle.
    x, y = 50, 50    # Starting coordinates (top-left corner)
    w, h = 200, 150  # Width and height of the cropped region

    # Crop the image using NumPy slicing
    cropped_image = image[y:y+h, x:x+w]

    # Convert images from BGR (OpenCV default) to RGB for proper display with matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    cropped_rgb = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB)

    # Display the original and cropped images using matplotlib
    plt.figure(figsize=(10, 5))

    # Original image
    plt.subplot(1, 2, 1)
    plt.imshow(image_rgb)
    plt.title("Original Image")
    plt.axis('off')

    # Cropped image
    plt.subplot(1, 2, 2)
    plt.imshow(cropped_rgb)
    plt.title("Cropped Image")
    plt.axis('off')

    plt.show()
```
