```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread("DATA/test1.jpg") # BGR

def img_display(image):
    cv2.imshow("outer window", image)
    cv2.waitKey()
    cv2.destroyAllWindows()
```

```python
img_display(image)
```

```python
image.shape
```

```python
plt.imshow(image) #RGB
```

```python
# We can cvtColor , to convert to grayscale
gray_image = cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)
```

```python
gray_image.shape
```

```python
plt.imshow(gray_image, cmap="gray")
```

```python
cv2.imshow("Grayscale", gray_image)
cv2.waitKey()
cv2.destroyAllWindows()
```
