## Histogram in OpenCV

A histogram is a graphical representation of the intensity distribution of an image. It shows how pixel intensity values are distributed across an image, which helps in image processing tasks like contrast adjustment, thresholding, and equalization.

### What is a Histogram?

* A histogram plots the number of pixels for each intensity value.
* In grayscale images, it shows pixel distribution from 0 (black) to 255 (white).
* In color images, separate histograms are plotted for the Blue (B), Green (G), and Red (R) channels.

```python
# Importing Dependencies
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the image
image = cv2.imread("DATA/DOGS.jpeg")
```

```python
# Checking Type
type(image)
```

```python
# Compute histogram
histogram = cv2.calcHist([image],
                        [1],
                        None,
                        [256],
                        [0,256])
```

```python
histogram
```

```python
# We plot the histogram , ravel() flatterns our image array
plt.hist(image.ravel(), 256,[0,256])
plt.show()
```

```python
# Viewing Saperate Color Channels
color =("b","g","r")

# We now saperate the color and plot each in the Histogram
for i, col in enumerate(color):
    histogram2 = cv2.calcHist([image],[i], None ,[256],[0,256])
    plt.plot(histogram2 , color=col)
    plt.xlim([0,256])

# Display the histogram
plt.title("Color Histogram (B, G, R)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
# Load the image in grayscale
image = cv2.imread('DATA/DOGS.jpeg', cv2.IMREAD_GRAYSCALE)  # Ensure it's grayscale

# Perform histogram equalization
equalized = cv2.equalizeHist(image)

# Display images
cv2.imshow("Original", image)
cv2.imshow("Equalized", equalized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
